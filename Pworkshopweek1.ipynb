```python
import numpy as np


empty_array = np.empty((2, 2))
print("Empty array (2x2):")
print(empty_array)


ones_array = np.ones((4, 2))
print("\nAll-one array (4x2):")
print(ones_array)


full_array = np.full((3, 3), 7)
print("\nFull array (3x3 filled with 7):")
print(full_array)


given_array = np.array([[1, 2], [3, 4]])
zeros_like_array = np.zeros_like(given_array)
print("\nZeros like given array:")
print(zeros_like_array)


ones_like_array = np.ones_like(given_array)
print("\nOnes like given array:")
print(ones_like_array)


new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("\nConverted numpy array:")
print(numpy_array)
```

```python
import numpy as np


A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])


A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)


AB = np.dot(A, B)
BA = np.dot(B, A)
AB_not_BA = not np.allclose(AB, BA)


AB_T = (AB).T
B_T_A_T = np.dot(B.T, A.T)
transpose_equal = np.allclose(AB_T, B_T_A_T)

print("A * A^-1 = \n", I)
print("\nAB = \n", AB)
print("BA = \n", BA)
print("AB != BA: ", AB_not_BA)
print("\n(AB)^T = \n", AB_T)
print("B^T A^T = \n", B_T_A_T)
print("Transpose equality: ", transpose_equal)

# Solve system of linear equations using inverse method
# Equations:
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

# Coefficient matrix
coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])

# Constant matrix
const_matrix = np.array([-1, -3, 9])

# Solution using inverse
solution = np.dot(np.linalg.inv(coeff_matrix), const_matrix)
print("\nSolution using inverse method:", solution)
```

```python
import numpy as np
import time

# -----------------------------
# Element-wise Addition
# -----------------------------
size = 1_000_000

# Using Python lists
list1 = list(range(size))
list2 = list(range(size))
start = time.time()
list_add = [list1[i] + list2[i] for i in range(size)]
end = time.time()
print("Python list addition time:", end - start, "seconds")

# Using NumPy arrays
arr1 = np.arange(size)
arr2 = np.arange(size)
start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy array addition time:", end - start, "seconds")

# -----------------------------
# Element-wise Multiplication
# -----------------------------
# Python lists
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(size)]
end = time.time()
print("Python list multiplication time:", end - start, "seconds")

# NumPy arrays
start = time.time()
arr_mul = arr1 * arr2
end = time.time()
print("NumPy array multiplication time:", end - start, "seconds")

# -----------------------------
# Dot Product
# -----------------------------
# Python lists
start = time.time()
dot_list = sum([list1[i] * list2[i] for i in range(size)])
end = time.time()
print("Python list dot product time:", end - start, "seconds")

# NumPy arrays
start = time.time()
dot_array = np.dot(arr1, arr2)
end = time.time()
print("NumPy dot product time:", end - start, "seconds")

# -----------------------------
# Matrix Multiplication (1000x1000)
# -----------------------------
matrix_size = 1000

# Python lists
matrix1 = [[i+j for j in range(matrix_size)] for i in range(matrix_size)]
matrix2 = [[i-j for j in range(matrix_size)] for i in range(matrix_size)]

start = time.time()
result = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(matrix_size))
           for j in range(matrix_size)] for i in range(matrix_size)]
end = time.time()
print("Python list matrix multiplication time:", end - start, "seconds")

# NumPy arrays
matrix1_np = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)
matrix2_np = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)

start = time.time()
result_np = np.dot(matrix1_np, matrix2_np)
end = time.time()
print("NumPy matrix multiplication time:", end - start, "seconds")
```
